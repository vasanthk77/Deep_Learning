```python
a=pd.read_csv("Churn_Modelling.csv")
```

```python
df=pd.DataFrame(a)
```

```python
df.head()
```

```python
a=pd.get_dummies(df['Geography'],drop_first=True)
a1=pd.get_dummies(df['Gender'],drop_first=True)
```

```python
df1=pd.concat([df,a,a1],axis='columns')
```

```python
df1.head()
```

```python
df1.drop(['Geography'],axis='columns',inplace=True)
```

```python

df1.drop(['CustomerId'],axis='columns',inplace=True)
df1.drop(['Surname'],axis='columns',inplace=True)
df1.drop(['RowNumber'],axis='columns',inplace=True)
```

```python
df1.head()
```

```python
X=df1.drop(['Exited'],axis='columns')
```

```python
y=df1.Exited
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
X_train
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LeakyReLU, ReLU, Dropout
```

```python
classifier=Sequential()
```

```python
##adding the layers
```

```python
##input layer
classifier.add(Dense(units=11,activation='relu'))
```

```python
##  adding the first hidden layer
classifier.add(Dense(units=7,activation='relu'))
```

```python
## second layer
classifier.add(Dense(units=6,activation='relu'))
```

```python
## output layer
classifier.add(Dense(units=1,activation='sigmoid'))
```

```python
classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model=classifier.fit(X_train,y_train,validation_split=0.33,batch_size=10,epochs=10)
```
